```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from math import sqrt

import shap
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping

```

```python
ticker = input("Enter stock ticker ( INFY.NS): ")
data = yf.download(ticker, start="2010-01-01")

```

```python
# Flatten MultiIndex columns (CRITICAL)
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Keep only Close as a TRUE Series
data = data[['Close']].copy()
data['Close'] = data['Close'].astype(float)
```

```python
data['MA20'] = data['Close'].rolling(20).mean()
data['MA50'] = data['Close'].rolling(50).mean()
```

```python
delta = data['Close'].diff()
gain = delta.clip(lower=0).rolling(14).mean()
loss = -delta.clip(upper=0).rolling(14).mean()
rs = gain / loss
data['RSI'] = 100 - (100 / (1 + rs))
```

```python
ema12 = data['Close'].ewm(span=12, adjust=False).mean()
ema26 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = ema12 - ema26
```

```python
data['Log_Return'] = np.log(data['Close'] / data['Close'].shift(1))
data['Volatility'] = data['Log_Return'].rolling(20).std()
data['Close_MA20_Diff'] = (
    data['Close'].values.reshape(-1) -
    data['MA20'].values.reshape(-1)
)
```

```python
data.dropna(inplace=True)
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['Close'], label="Close Price")
plt.plot(data['MA20'], label="MA20")
plt.plot(data['MA50'], label="MA50")
plt.title(f"{ticker} Price with Moving Averages")
plt.legend()
plt.show()
```

```python
def create_dataset(dataset, look_back):
    X, y = [], []
    for i in range(look_back, len(dataset)):
        X.append(dataset[i-look_back:i])
        y.append(dataset[i, 0])  # Close price
    return np.array(X), np.array(y)

def build_lstm(input_shape, units=64):
    model = Sequential()
    model.add(LSTM(units, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.3))
    model.add(LSTM(units))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
look_back = 60
train_size = int(len(data) * 0.8)

train_raw = data.iloc[:train_size]
test_raw = data.iloc[train_size:]

scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_raw)
test_scaled = scaler.transform(test_raw)

X_train, y_train = create_dataset(train_scaled, look_back)
test_input = np.vstack((train_scaled[-look_back:], test_scaled))
X_test, y_test = create_dataset(test_input, look_back)
```

```python
model = build_lstm((X_train.shape[1], X_train.shape[2]))
es = EarlyStopping(patience=6, restore_best_weights=True)

model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1,
    callbacks=[es],
    verbose=0
)

preds = model.predict(X_test, verbose=0)

preds_inv = scaler.inverse_transform(
    np.hstack((preds, np.zeros((len(preds), train_scaled.shape[1]-1))))
)[:, 0]

y_inv = scaler.inverse_transform(
    np.hstack((y_test.reshape(-1,1), np.zeros((len(y_test), train_scaled.shape[1]-1))))
)[:, 0]


print("Base LSTM RMSE:", sqrt(mean_squared_error(y_inv, preds_inv)))
```

```python
X_shap = train_raw.drop(columns=['Close'])
y_shap = train_raw['Close']

surrogate = RandomForestRegressor(n_estimators=200, random_state=42)
surrogate.fit(X_shap, y_shap)

explainer = shap.Explainer(surrogate, X_shap)
shap_values = explainer(X_shap)

# SHAP plots
shap.summary_plot(shap_values, X_shap, plot_type="bar")
shap.summary_plot(shap_values, X_shap)
```

```python
selected_features = [
    'Close', 'RSI', 'MACD', 'Volatility', 'MA20'
]

data_reduced = data[selected_features]
```

```python
train_red = data_reduced.iloc[:train_size]
test_red = data_reduced.iloc[train_size:]

scaler2 = MinMaxScaler()
train_scaled2 = scaler2.fit_transform(train_red)
test_scaled2 = scaler2.transform(test_red)

X_train2, y_train2 = create_dataset(train_scaled2, look_back)
test_input2 = np.vstack((train_scaled2[-look_back:], test_scaled2))
X_test2, y_test2 = create_dataset(test_input2, look_back)

model2 = build_lstm((X_train2.shape[1], X_train2.shape[2]), units=32)

model2.fit(
    X_train2, y_train2,
    epochs=30,
    batch_size=32,
    validation_split=0.1,
    callbacks=[es],
    verbose=0
)

preds2 = model2.predict(X_test2, verbose=0)

preds2_inv = scaler2.inverse_transform(
    np.hstack((preds2, np.zeros((len(preds2), train_scaled2.shape[1]-1))))
)[:, 0]

y2_inv = scaler2.inverse_transform(
    np.hstack((y_test2.reshape(-1,1), np.zeros((len(y_test2), train_scaled2.shape[1]-1))))
)[:, 0]

print("SHAP-Selected LSTM RMSE:", sqrt(mean_squared_error(y2_inv, preds2_inv)))
```

```python
plt.figure(figsize=(12,5))
plt.plot(y2_inv, label='Actual')
plt.plot(preds2_inv, label='Predicted (SHAP Selected)')
plt.title("Improved LSTM using SHAP-Based Feature Selection")
plt.legend()
plt.show()
```
